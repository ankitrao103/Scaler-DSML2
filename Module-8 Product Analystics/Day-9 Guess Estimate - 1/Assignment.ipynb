# Q1. Use of Guesstimates

Guess estimates are generally meant to… (Select the most appropriate one)
- Check how good you’re with mind calculations
- Find out if you can guess the exact numbers
- See if you can derive a structured approach
- None of these

**Answer:** See if you can derive a structured approach

**Explanation:** A guess estimate is an approximate calculation or estimation made without using precise or accurate figures. It involves using your knowledge, experience, and judgment to arrive at an educated guess or rough estimate.

The purpose of guess estimates is to develop a structured approach for solving a problem, identifying potential solutions, and evaluating different options.
It is a useful skill in many fields, including business, finance, engineering, and science.

# Q3. Revenue for Gaming Event 1

Assuming that…

  1. Each entry pass costs around Rs. 500/-
  2. The no. of active players on the platform is 15k
  3. Only 22% of the total players are paid users
  
What will be the total revenue generated by selling entry passes for the gaming event?

**Answer:** 16.5 lakhs

**Solution:**

Total revenue = No. of entry passes sold * Cost of each entry pass

No. of entry passes sold = No. of paid users i.e., 22% of 15k
(22/100) * 15,000 = 3,300

Given,
Cost of each entry pass = Rs. 500/-

Therefore, Total revenue generated = 3,300 * 500 = 1650000 = **16.5 lakhs**

# Q4. Revenue for Gaming Event 2

Cost of each entry pass = Rs. 1500/-

Therefore, Total revenue generated = 1,250 * 1500 = 1875000 = **18.75 lakhs**

# Q5. Revenue by Ads Hosting

Assuming that…

  1. Every day, 2 advertisements are shown to the free players whenever they open the game for the first time.
  2. The first ad is of 30-sec duration and the second one is of 45-sec duration.
  3. Hosting a 30-sec ad on the platform costs around 30k per ad, and a 45-sec ad around 50k per ad, to the partner companies.

What will be the monthly revenue generated by hosting ads on the platform?

**Note:** Assume the no. of days in a month to be 30.

**Answer:** 24 lakhs/month

**Solution:**

Total monthly revenue =
Revenue generated per month by the 30-sec ads + Revenue generated per month by the 45-sec ads

Let us represent this equation by rev_mon = rev_30 + rev_45

If we disregard the ad duration, then
Monthly revenue = No. of ads per day * No. of days in a month * Cost of hosting 1 ad

Given,
No. of days in a month = 30 (As per the given assumption)
Cost of hosting a 30-sec ad = 30k
Cost of hosting a 30-sec ad = 50k

rev_30 = 1 * 30 * 30k = 900k
rev_45 = 1 * 30 * 50k = 1500k

Therefore,
Total monthly revenue generated (rev_mon) = rev_30 + rev_45 = 900k + 1500k = 2400000 = **24 lakhs**

# Q6. Budget for Self Promotion

Assuming that…

  1. The gaming company spends around 25k per day on advertising its application via YouTube and other such channels.
  2. They’ve hired some celebs as well for self-promotion, for whom they pay a sum of 12 lakhs per month.
  3. Every month, the company organizes a fan event where they host an audience of 5k and distributes free gaming merchandise to them.
  4. The manufacturing cost of each of these merchandise is Rs. 120/-
What should be the total monthly budget allocated by the company for self-promotion?

**Note:** Assume the no. of days in a month to be 30.

**Answer:** 25.5 lakhs/month

**Solution:**
Total monthly expenditure = Expenditure on promotions via ad per month + Celebrity fee per month + Manufacturing cost of merchandise distributed each month

Expenditure on promotions via ad = 25k (per day)
Hence, for 1 month the expenditure will be = 25,000 * 30 = 750000 = 7.5 lakhs

Given,
Celebrity fee = 12 lakhs per month

Total expenditure on merchandise manufacturing = No. of people present in the fan event * No. of merchandise given to each one of them * Manufacturing cost of 1 merchandise
= 5,000 * 1 * 120 = 600000 = 6 lakhs per month

Therefore, Total monthly expenditure = 7.5 lakhs + 12 lakhs + 6 lakhs = 2550000 = **25.5 lakhs**
