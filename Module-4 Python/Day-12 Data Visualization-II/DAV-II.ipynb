```python

#! Top 3 Data -> Top 3 Publishers and Top 3 Genres adn Top 3 Platforms 
```

```python
#* Top 3 Publishers
top3_publishers = data["Publisher"].value_counts().index[:3]
top3_publishers
```

```python
top3_publishers[0]
```

```python
# Top 3 Genres
top3_genres = data["Genre"].value_counts().index[:3]
top3_genres
```

```python
# Top 3 Platforms
top3_platforms = data["Platform"].value_counts().index[:3]
top3_platforms
```

```python
top3_data = data.loc[
  (data["Publisher"].isin(top3_publishers)) &
  (data["Genre"].isin(top3_genres)) &
  (data["Platform"].isin(top3_platforms))
]
top3_data
```

# **SubPlots**

```python
sns.scatterplot(
  x = top3_data["NA_Sales"], 
  y = top3_data["EU_Sales"]
)

sns.scatterplot(
  x = top3_data["NA_Sales"], 
  y = top3_data["JP_Sales"]
)

sns.scatterplot(
  x = top3_data["NA_Sales"], 
  y = top3_data["Other_Sales"]
)

#! I donot wanna merge these all three but want 3 diff plots
plt.show()
```

```python
plt.figure(figsize = (15,5)) # Width, Height
#! Subplots
plt.subplot(1,3,1) # 1 row, 3 columns, 1st plot
sns.scatterplot(
  x = top3_data["NA_Sales"], 
  y = top3_data["EU_Sales"]
)

plt.subplot(1,3,2) # 1 row, 3 columns, 2st plot
sns.scatterplot(
  x = top3_data["NA_Sales"], 
  y = top3_data["JP_Sales"]
)

plt.subplot(1,3,3) # 1 row, 3 columns, 3st plot
sns.scatterplot(
  x = top3_data["NA_Sales"], 
  y = top3_data["Other_Sales"]
)

plt.show()
```

```python
plt.figure(figsize = (15,5)) # Width, Height
#! Subplots
plt.subplot(2,3,1) # 1 row, 3 columns, 1st plot
sns.scatterplot(
  x = top3_data["NA_Sales"], 
  y = top3_data["EU_Sales"]
)

plt.subplot(2,3,4) # 1 row, 3 columns, 2st plot
sns.scatterplot(
  x = top3_data["NA_Sales"], 
  y = top3_data["JP_Sales"]
)

plt.subplot(2,3,6) # 1 row, 3 columns, 3st plot
sns.scatterplot(
  x = top3_data["NA_Sales"], 
  y = top3_data["Other_Sales"]
)

plt.show()
```

## **Another way to create grid for subplots**

```python

#! Using Axes

fig, ax = plt.subplots(2,2, figsize = (15,10)) # Width, Height
ax[0,0].scatter(x = top3_data["NA_Sales"], y = top3_data["EU_Sales"])
ax[0,1].scatter(x = top3_data["NA_Sales"], y = top3_data["JP_Sales"])
ax[1,0].scatter(x = top3_data["NA_Sales"], y = top3_data["Other_Sales"])
ax[1,1].scatter(x = top3_data["NA_Sales"], y = top3_data["Global_Sales"])
```

```python
plt.figure(figsize=(20,12)).suptitle("NA Sales vs regions",fontsize=20)

# Using a 2x3 subplot
plt.subplot(2, 3, 1)
sns.scatterplot(x='NA_Sales', y='EU_Sales', data=top3_data)
plt.title('NA vs EU Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('EU', fontsize=12)

plt.subplot(2, 3, 3)
sns.scatterplot(x='NA_Sales', y='JP_Sales', data=top3_data, color='red')
plt.title('NA vs JP Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('JP', fontsize=12)

plt.subplot(2, 3, 4)
sns.scatterplot(x='NA_Sales', y='Other_Sales', data=top3_data, color='green')
plt.title('NA vs Other Region Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('Other', fontsize=12)

plt.subplot(2, 3, 6)
sns.scatterplot(x='NA_Sales', y='Global_Sales', data=top3_data, color='orange')
plt.title('NA vs Global Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('Global', fontsize=12)

plt.subplot(1, 3, 2)
sns.countplot(x = "Publisher", data = top3_data)
plt.title("Count of Games by each Publisher")

plt.show()
```
