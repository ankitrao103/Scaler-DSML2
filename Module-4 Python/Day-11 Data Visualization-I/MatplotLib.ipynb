# Anatomy

## How to choose Right chart?
 - How many variables are Involed!
   - 1 Variable - Univariate Ananlysis
   - 2 Variables - Bivariate Analysis
   - 2 or more Variables - Multivariate Analysis
- What type of variables are involved?
  - Categorical - Bar Chart, Pie Chart
  - Continuous - Histogram, Box Plot, Scatter Plot
