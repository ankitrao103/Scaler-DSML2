```python
'''
Risky Movies:
  - Budget --> Avg Revenue of the Directior and then subtract Avg Revenue from the budget
'''
```

```python
def risky_func(x):
  x["risky"] = (x["budget"] - x["revenue"].mean()) >=0
  return x
```

```python
data_risky = data.groupby('director_name').apply(risky_func)
```
