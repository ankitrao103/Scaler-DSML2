```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('movies.csv', index_col=0)
directors = pd.read_csv('directors.csv', index_col=0)
```

```python
movies
```

```python
directors
```

```python
data = movies.merge(directors, how='left', left_on='director_id', right_on='id')
```

```python
data.head()
```

```python
data.drop(["director_id", "id_y"], axis=1, inplace=True)
```

```python
data.head()
```

# Which Director is the most Productive?

```python
'''
1. group by director_name
2. Count the title
3. Sort in descending order
'''
```

```python
data.groupby(["director_name"])["title"].count()
```

```python
data.groupby(["director_name"])["title"].count().sort_values(ascending=False)
```

```python
# Steven Spielberg has the most productive director
```

```python
'''
Analysis 2: Total movies director per year 

1. Calculate min and max for every director
2. Calculate the year active
3. Divide title count by years active and sort in a descending order
'''
```

```python
# Step-1
data_agg = data.groupby(["director_name"])[["year","title"]].aggregate({
  "year": ["min", "max"], 
  "title":"count"
})
```

```python
data_agg
```

```python
data_agg.columns
```

```python
data_agg["year"]["max"]
```

```python
data_agg["year"]["min"]
```

```python
data_agg["title"]
```

```python
data_agg["title"]["count"]
```

```python
data_agg = data.groupby(["director_name"])[["year","title"]].aggregate(
  year_min = ("year", "min"),
  year_max = ("year", "max"),
  title_count = ("title", "count")   
)
```

```python
data_agg
```

```python
data_agg.reset_index(inplace=True)
```

```python
data_agg
```

```python
data_agg['years_active'] = data_agg['year_max'] - data_agg['year_min']
data_agg
```

```python
data_agg['movies_per_year'] = data_agg['title_count'] / data_agg['years_active']
```

```python
data_agg
```

```python
data_agg.sort_values("movies_per_year", ascending=False)
```

### Melting(Converting column into Rows)

```python
pd.melt(pfizer, id_vars=['Date', 'Drug_Name', 'Parameter'])
```

```python
data_melt = pd.melt(
  pfizer,
  id_vars=['Date', 'Drug_Name', 'Parameter'],
  var_name='Time',
  value_name='Reading'
)
```

```python
data_melt
```

```python
data_melt.pivot(
  index = ['Date', 'Drug_Name', 'Parameter'],
  columns='Time',
  values='Reading'
).reset_index()
```

```python
# Index in df.pivot adn id_vars in pd,melt play similar roles! 
```

```python
'''
Idea format for this data
Date | Time | Drug_Name | Pressure | Temperature 
Pressure and Temperature are the coverted from Rows to Columns
'''
```

```python
data_melt
```

```python
data_tidy = data_melt.pivot(
  index=['Date','Time','Drug_Name'],
  columns= "Parameter",
  values='Reading'
)
```

## **Pivot Tables**

```python
# Daywise Mean value of Temperature for each Drug
# Can be Possible using groupby and Aggregate
```

```python
# WIth PIVOT_TABLE
```

```python
data_tidy.pivot(
  index=["Drug_Name"],
  columns="Date",
  values="Temperature"
)
```

```python
# Pivot table is also using groupby and aggregate internally
pd.pivot_table(
  data_tidy,
  index=["Drug_Name"],
  columns="Date",
  values="Temperature",
  aggfunc= np.mean
).reset_index()
```

```python
# This Syntax also works
data_tidy.pivot_table(
  index=["Drug_Name"],
  columns="Date",
  values=["Temperature", "Pressure"],
  aggfunc= np.max
).reset_index()
```

```python
# This Syntax also works
data_tidy.pivot_table(
  index=["Drug_Name"],
  columns="Date",
  values=["Temperature", "Pressure"],
  aggfunc= [np.max, np.min]
).reset_index()
```
