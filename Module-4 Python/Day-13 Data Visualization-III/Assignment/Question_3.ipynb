```python
import seaborn as sns
import numpy as np
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.countplot(x='survived', data=titanic, hue='class')
```

```python
df = titanic[titanic['class'] == 'Third']
df['fare'].mean(), df['fare'].median()
```

```python
sns.barplot(
  x = 'class',
  y = 'fare',
  data = titanic,
  hue = 'survived',
  estimator = np.mean
)
```

```python
df = titanic[titanic['class'] == 'Second']
```

```python
sns.violinplot(
  x = 'class',
  y = 'fare',
  data = df,
  hue = 'survived',
)
```
