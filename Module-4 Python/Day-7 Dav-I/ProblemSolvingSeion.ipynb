# **Image Manipulation**

```python
import numpy as np
import matplotlib.pyplot as plt

!gdown 1UL59W-QUysWJ_QtJMx0Wdj30g0Qcu8LR
birds = plt.imread("birds.jpeg")

birds
```

```python
#displaying the image from the array
plt.imshow(birds, cmap="gray")
```

```python
#Display a horizontally inverted image (mirror image)
plt.imshow(birds[::,::-1], cmap="gray")
```

```python
#Display a VERTICALLY inverted image (vetical mirror image)
plt.imshow(birds[::-1,::], cmap="gray")
```

```python
# Inverted Horizontally and Vertically
plt.imshow(birds[::-1,::-1], cmap="gray")
```

```python
#transpose
plt.imshow(birds.T, cmap="gray")
```

```python
# Rotation by 90 degrees
#transpose and then Rev columns
plt.imshow(birds.T[::,::-1], cmap="gray")
```
